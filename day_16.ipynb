```python
import pandas as pd
import numpy as np 

input_test = r""".|...\....
|.-.\.....
.....|-...
........|.
..........
.........\
..../.\\..
.-.-/..|..
.|....-|.\
..//.|...."""

with open('data/day_16.txt') as txtfile:
    input_day = txtfile.read()

def get_data(prod=0):
    input_data = input_day if prod else input_test
    return input_data

def coord_is_valid(row, col):
    return (row >= 0) and (row<=size) and (col >= 0) and (col<=size)

def mirror_up():
    new_pos = [row - 1, col + sentido]
    if coord_is_valid(new_pos[0], new_pos[1]):
        paths["last_path"] += 1
        paths[paths["last_path"]] = {}
        paths[paths["last_path"]]["path"] = [new_pos]
        if (df.loc[tuple(new_pos)] == "|") or (df.loc[tuple(new_pos)] == "."):
            paths[paths["last_path"]]["sentido"] = 1
            paths[paths["last_path"]]["direction"] = not direction
        elif df.loc[tuple(new_pos)] == "/":
            paths[paths["last_path"]]["sentido"] = 1
            paths[paths["last_path"]]["direction"] = direction
        elif df.loc[tuple(new_pos)] == "\\":
            paths[paths["last_path"]]["sentido"] = -1
            paths[paths["last_path"]]["direction"] = direction


def mirror_down():
    new_pos = [row + 1, col + sentido]
    if coord_is_valid(new_pos[0], new_pos[1]):
        paths["last_path"] += 1
        paths[paths["last_path"]] = {}
        paths[paths["last_path"]]["path"] = [new_pos]
        paths[paths["last_path"]]["sentido"] = 1
        paths[paths["last_path"]]["direction"] = not direction

def mirror_left():
    new_pos = [row + sentido, col - 1]
    if coord_is_valid(new_pos[0], new_pos[1]):
        paths["last_path"] += 1
        paths[paths["last_path"]] = {}
        paths[paths["last_path"]]["path"] = [new_pos]
        paths[paths["last_path"]]["sentido"] = -1
        paths[paths["last_path"]]["direction"] = not direction

def mirror_right():
    new_pos = [row + sentido, col + 1]
    if coord_is_valid(new_pos[0], new_pos[1]):
        paths["last_path"] += 1
        paths[paths["last_path"]] = {}
        paths[paths["last_path"]]["path"] = [new_pos]
        if (df.loc[tuple(new_pos)] == "-") or (df.loc[tuple(new_pos)] == "."):
            paths[paths["last_path"]]["sentido"] = 1
            paths[paths["last_path"]]["direction"] = not direction
        elif df.loc[tuple(new_pos)] == "/":
            paths[paths["last_path"]]["sentido"] = - 1
            paths[paths["last_path"]]["direction"] = direction
        elif df.loc[tuple(new_pos)] == "\\":
            paths[paths["last_path"]]["sentido"] = 1
            paths[paths["last_path"]]["direction"] = direction

def move_next_col():
    new_pos = [row, col + sentido]
    if coord_is_valid(new_pos[0], new_pos[1]):
        paths[current_path]["path"].append(new_pos)
        return tuple(new_pos)

def move_next_row():
    new_pos = [row + sentido, col]
    if coord_is_valid(new_pos[0], new_pos[1]):
        paths[current_path]["path"].append(new_pos)
        return tuple(new_pos)
```

```python
data = get_data(0)
df = pd.DataFrame([list(l.strip()) for l in data.splitlines()])
df
```

```python
paths = {"last_path": 0, 0: {"sentido": 1, "direction": False, "path": [[7, 2]]}}
current_path = 0
visited = []
visited_paths = []
size = df.shape[0] - 1


while current_path not in visited_paths:
    row, col = paths[current_path]["path"][0]
    direction = paths[current_path]["direction"]
    sentido = paths[current_path]["sentido"]
    old_row, old_col = row, col
    nx = row, col + sentido
    if direction == False:
        if coord_is_valid(nx[0], nx[1]):
            while (df.loc[nx]=="-") or (df.loc[nx]=="."):
                nx = move_next_col()
                if not bool(nx) or ((row, col, direction) in visited):
                    visited_paths.append(current_path)
                    break
                else:
                    visited.append((row, col, direction))
                    old_row, old_col = row, col
                    row, col = nx
            else:
                row, col = old_row, old_col
                if (df.loc[nx]=="\\"):
                    if sentido == 1:
                        mirror_down()
                    else:
                        mirror_up()
                elif (df.loc[nx]=="/"):
                    if sentido == -1:
                        mirror_down()
                    else:
                        mirror_up()
                elif df.loc[nx]=="|":
                    mirror_up()
                    mirror_down()
                visited.append((row, col, direction))
                visited_paths.append(current_path)
                current_path += 1
        else:
            current_path += 1
    else:
        nx = row + sentido, col
        if coord_is_valid(nx[0], nx[1]):
            while (df.loc[nx]=="|") or (df.loc[nx]=="."):
                nx = move_next_row()
                if not bool(nx) or ((row, col, direction) in visited):
                    visited_paths.append(current_path)
                    break
                else:
                    visited.append((row, col, direction))
                    old_row, old_col = row, col
                    row, col = nx
            else:
                row, col = old_row, old_col
                if (df.loc[nx]=="\\"):
                    if sentido == 1:
                        mirror_right()
                    else:
                        mirror_left()
                elif (df.loc[nx]=="/"):
                    if sentido == -1:
                        mirror_right()
                    else:
                        mirror_left()
                elif df.loc[nx]=="-":
                    mirror_right()
                    mirror_left()
                visited.append((row, col, direction))
                visited_paths.append(current_path)
                current_path += 1
        else:
            current_path += 1

paths
```
