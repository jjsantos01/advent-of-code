```python
import pandas as pd
import numpy as np 

input_test = """O....#....
O.OO#....#
.....##...
OO.#O....O
.O.....O#.
O.#..O.#.#
..O..#O..O
.......O..
#....###..
#OO..#...."""

with open('data/day_14.txt') as txtfile:
    input_day = txtfile.read()

def get_data(prod=0):
    input_data = input_day if prod else input_test
    return input_data
```

```python
data = get_data(1)
df = pd.DataFrame([list(l) for l in data.splitlines()])
df_new = df.copy()
df_new
```

# Part I

```python
def make_tilt(df_new):
    for i in range(df_new.shape[0]-1, 0, -1):
        indexes = df_new.iloc[i-1].eq(".") & df_new.iloc[i].eq("O")
        while indexes.sum():
            i_values = df_new.iloc[i].copy()
            n_values = df_new.iloc[i-1].copy()
            df_new.iloc[i-1] = np.where(indexes, i_values, n_values)
            df_new.iloc[i] = np.where(indexes, n_values, i_values)
            if (i < df_new.shape[0]-1):
                indexes = df_new.iloc[i].eq(".") & df_new.iloc[i+1].eq("O")
                if indexes.sum():
                    i+=1
            else:
                break
    return df_new

out = make_tilt(df_new)
solution = out.eq("O").sum(axis=1).multiply(np.arange(len(out), 0, -1)).sum()
print(solution)
```

# Part 2

```python
out = df.copy()
history = []
for c in range(200):
    for _ in range(4):
        out = make_tilt(out).T.iloc[:,::-1]
    history.append(out.copy())
    
    
def get_first_repeated_cycle(history):
    for i, first in enumerate(history):
        for j, h in enumerate(history):
            if i!=j:
                if h.eq(first).all().all():
                    return i, j
                
i, j = get_first_repeated_cycle(history)
delta = j-i
cycles = 1_000_000_000
steps = (cycles - (i+1)) // delta
lefting = cycles - (i + 1 + steps * delta)
ps = i + lefting
out = history[ps]
solution = out.eq("O").sum(axis=1).multiply(np.arange(len(out), 0, -1)).sum()
print(solution)
```
