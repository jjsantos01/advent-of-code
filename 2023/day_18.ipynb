```python
from shapely import Polygon

input_test = r"""R 6 (#70c710)
D 5 (#0dc571)
L 2 (#5713f0)
D 2 (#d2c081)
R 2 (#59c680)
D 2 (#411b91)
L 5 (#8ceee2)
U 2 (#caa173)
L 1 (#1b58a2)
U 2 (#caa171)
R 2 (#7807d2)
U 3 (#a77fa3)
L 2 (#015232)
U 2 (#7a21e3)"""

with open('data/day_18.txt') as txtfile:
    input_day = txtfile.read()

def get_data(prod=0):
    input_data = input_day if prod else input_test
    return input_data
```

# Part 1

```python
data = get_data(1)
lines = [[j[0], int(j[1]), j[2][1:-1]] for j in [i.split() for i in data.splitlines()]]
start = (0, 0)
history = []
for l in lines:
    d, n, _ = l
    row, col = start
    next_row, next_col = start
    if d == "R":
        next_col += n
        history += [(row, col), (row, next_col)]
    elif d == "L":
        next_col -= n
        history += [(row, col), (row, next_col)]
    elif d == "D":
        next_row += n
        history += [(row, col), (next_row, col)]
    else:
        next_row -= n
        history += [(row, col), (next_row, col)]
    start = (next_row, next_col)
poly = Polygon(history)
print(poly.area + (poly.length / 2) + 1)
```

# Part II

```python
data = get_data(1)
lines = [[j[0], int(j[1]), j[2][1:-1]] for j in [i.split() for i in data.splitlines()]]
start = (0, 0)
history = []
for l in lines:
    _, _, color = l
    n = int(color[1:6], 16)
    d = int(color[6], 16)
    row, col = start
    next_row, next_col = start
    if d == 0:
        next_col += n
        history += [(row, col), (row, next_col)]
    elif d == 2:
        next_col -= n
        history += [(row, col), (row, next_col)]
    elif d == 1:
        next_row += n
        history += [(row, col), (next_row, col)]
    else:
        next_row -= n
        history += [(row, col), (next_row, col)]
    start = (next_row, next_col)
poly = Polygon(history)
print(poly.area + (poly.length / 2) + 1)
```
