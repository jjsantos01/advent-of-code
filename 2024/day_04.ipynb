```python
import re
import pandas as pd
import numpy as np

input_test = """MMMSXXMASM
MSAMXMSMSA
AMXSXMAAMM
MSAMASMSMX
XMASAMXAMM
XXAMMXXAMA
SMSMSASXSS
SAXAMASAAA
MAMMMXMMMM
MXMXAXMASX"""

with open('data/day_04.txt') as txtfile:
    input_day = txtfile.read()
    
PROD = 1
input_data = input_day if PROD else input_test
print(input_data[0:200])
```

```python
df = pd.DataFrame([list(line) for line in input_data.split()])
trans = "\n".join(["".join(l) for l in pd.DataFrame([list(line) for line in input_data.split()]).T.values.tolist()])
df
```

# PART 1

```python
re_xmas = re.compile(r"(?=(XMAS|SAMX))")
xmas = re_xmas.findall(input_data) # horizontal search
xmas += re_xmas.findall(trans) # vertical search
index = len(df) - 4
for i in range(-index, index + 1):
    xmas += re_xmas.findall("".join(np.diag(df, k=i))) # diag 1 search
    xmas += re_xmas.findall("".join(np.diag(df.iloc[:,::-1], k=i))) # diag 2 search
len(xmas)
```

# PART 2

```python
n = 0
for r, c in df.stack().loc[lambda x: x.eq("A")].index:
    if (r > 0) and (r < len(df) - 1) and (c > 0) and (c < len(df) - 1):
        diag_1 = (df.loc[r-1, c-1] + df.loc[r+1, c+1]) in ["MS", "SM"]
        diag_2 = (df.loc[r-1, c+1] + df.loc[r+1, c-1]) in ["MS", "SM"]
        n += (diag_1 and diag_2)
n
```
