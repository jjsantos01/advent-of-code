```python
def get_neighbors(pos):
    path = {pos: []}
    pos_row, pos_col = np.array(pos)
    (l, r), (t, b) = maze_m.loc[pos_row, pos_col]
    if l & (pos_col >= 1):
        neigh_l = maze_m.loc[pos_row, pos_col - 1]
        if neigh_l[0][1]:
            path[pos_row, pos_col].append((pos_row, pos_col - 1))
    if r & (pos_col < maze_m.shape[1] - 1):
        neigh_r = maze_m.loc[pos_row, pos_col + 1]
        if neigh_r[0][0]:
            path[pos_row, pos_col].append((pos_row, pos_col + 1))
    if t & (pos_row >= 1):
        neigh_t = maze_m.loc[pos_row - 1, pos_col]
        if neigh_t[1][1]:
            path[pos_row, pos_col].append((pos_row - 1, pos_col))
    if b & (pos_row < maze_m.shape[0] - 1):
        neigh_b = maze_m.loc[pos_row + 1, pos_col]
        if neigh_b[1][0]:
            path[pos_row, pos_col].append((pos_row + 1, pos_col))

    return path
```

# Parte 1

```python
PROD = 1
test = 5
input_data = input_day if PROD else eval(f"input_test_{test}")

maze = pd.DataFrame([list(i) for i in input_data.splitlines()])
directions = {
    "-": np.array([(1, 1), (0, 0)]),
    "|": np.array([(0, 0), (1, 1)]),
    "L": np.array([(0, 1), (1, 0)]),
    "J": np.array([(1, 0), (1, 0)]),
    "7": np.array([(1, 0), (0, 1)]),
    "F": np.array([(0, 1), (0, 1)]),
    ".": np.array([(0, 0), (0, 0)]),
}
if PROD:
    directions["S"] = directions["J"]
else:
    if test==5:
        directions["S"] = directions["7"]
    else:
        directions["S"] = directions["F"]

maze_m = maze.applymap(lambda x: directions.get(x))

start = maze.eq("S").sum(axis=1).argmax(), maze.eq("S").sum().argmax()
neighbors = {}
for row in maze_m.index:
    for col in maze_m.columns:
        pos = row, col
        neighbors.update(get_neighbors(pos))

neigs = neighbors[start]
explorados = [start]
while (len(neigs)>0):
    prox = [n for n in neigs if n not in explorados]
    if prox:
        explorados.append(prox[0])
        neigs = neighbors[prox[0]]
    else:
        if start in neigs:
            explorados.append(start)
        break

assert explorados[0] == explorados[-1]
print((len(explorados)-1)/2)
```

```python
maze_full = pd.DataFrame(np.zeros_like(maze))
for row, col in zip(np.array(explorados)[:, 0], np.array(explorados)[:, 1]):
    maze_full.loc[row, col] = 1
maze_full
```

```python
import shapely
import re
    
shell = shapely.Polygon([(i[1], maze.shape[0]-i[0]) for i in explorados])
maze_string = maze_full.astype(str).apply(list, axis=1).str.join("").tolist()
puntos = {}
for p, l in enumerate(maze_string):
    matches = [a for a in re.finditer(r"(?<=1)0+(?=1)", l)]
    for m in matches:
        punto = []
        span = m.span()
        punto.append((span[0], maze.shape[0]-p))
        punto.append((span[1]-1, maze.shape[0]-p))
        puntos[p, span] = {"p": punto, "m": m}

solution = 0
for p in puntos.values():
    lenght = p["m"].end()-p["m"].start()
    if lenght == 1:
        poly_zeros = shapely.Point(p["p"][0])
    else:
        poly_zeros = shapely.LineString(p["p"])
    if poly_zeros.intersects(shell):
        solution += p["m"].end()-p["m"].start()
print(solution)
```
