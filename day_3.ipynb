```python
import pandas as pd
import numpy as np

with open('data/day_3.txt') as txtfile:
    input_day3 = txtfile.read()#.splitlines()
```

# Part 1

```python
df = pd.DataFrame([list(i) for i in input_day3.splitlines()])
df
```

```python
import re
ones = df.applymap(lambda x: not bool(re.search(r"\w|\.",x))).astype(int)
ones
```

```python
one_mask = (ones + ones.shift(-1).fillna(0) + ones.shift(1).fillna(0))\
    .pipe(lambda x: x + x.shift(-1,axis=1).fillna(0) + x.shift(1, axis=1).fillna(0)).astype(int)

one_mask
```

```python
numbers_one = df.applymap(lambda x: int(x in '0123456789'))
numbers_one
```

```python
twos = numbers_one * one_mask + numbers_one
twos
```

```python
numbers = df.applymap(lambda x: x * int(x in '0123456789') if x in '0123456789' else 0).astype(int)
numbers
```

```python
final_numbers = twos.astype(str).apply(lambda x: ''.join(x), axis=1)\
    .str.replace('0', '-')\
    .str.replace('\d*2\d*', repl=lambda x: '2' * (x.span()[1] - x.span()[0]), regex=True)\
    .str.replace('-', '1')\
    .apply(lambda x: pd.Series(list(x)))\
    .astype(int)\
    .replace({1: "", 2: 1})\
    .mul(numbers)\
    .replace("", "-")\
    .astype(str).apply(lambda x: ''.join(x), axis=1)\
    .str.findall("\d+")\
    .sum()
final_numbers
sum([int(x) for x in final_numbers])
```

```python
stars = df.applymap(lambda x: x=='*').astype(int)
stars_mask = (stars + stars.shift(-1).fillna(0) + stars.shift(1).fillna(0))\
    .pipe(lambda x: x + x.shift(-1,axis=1).fillna(0) + x.shift(1, axis=1).fillna(0)).astype(int)

stars_mask
```

```python
stars_id = pd.DataFrame([[f'{r}+{c}' for c in range(df.shape[0])] for r in range(df.shape[1])]) * stars
stars_mask_id = (stars_id + stars_id.shift(-1).fillna('') + stars_id.shift(1).fillna(''))\
    .pipe(lambda x: x + x.shift(-1,axis=1).fillna('') + x.shift(1, axis=1).fillna(''))

stars_mask_id
```

```python
twos_star = numbers_one * stars_mask + numbers_one
twos_star
```

```python
final_numbers = twos_star.astype(str).apply(lambda x: ''.join(x), axis=1)\
    .str.replace('0', '-')\
    .to_frame('name')\
    .apply(lambda x: re.sub('\d*2\d*',string=x["name"],
                            repl=lambda y: f"({max(stars_mask_id.loc[x.name, y.span()[0]], stars_mask_id.loc[x.name, y.span()[1]-1])}:{''.join(df.loc[x.name, y.span()[0]:y.span()[1]-1].tolist())})"),
           axis=1)\
    .str.findall("\([\d\+:]+\)")\
    .sum()

dicc_numbers = {}
for n in final_numbers:
    key = n.split(':')[0][1:]
    value = int(n.split(':')[1][:-1])
    if key not in dicc_numbers:
        dicc_numbers[key] = [value]
    else:
        dicc_numbers[key].append(value)

prod = 0
for val in dicc_numbers.values():
    if len(val)==2:
        prod += val[0] * val[1]

prod
```
